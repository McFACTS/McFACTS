```python
import itertools
import os

import numpy as np
import matplotlib.pyplot as plt

from mcfacts.inputs.settings_manager import SettingsManager
from mcfacts.objects.snapshot import TxtSnapshotHandler
```

```python
settings = SettingsManager({
    "verbose": False,
    "override_files": True,
    "save_state": True,
    "save_each_timestep": True
})

file_path = "../runs"
plots_dir = "../runs/plots"

snapshot_handler = TxtSnapshotHandler(settings)

dirs = os.listdir(file_path)

prograde_ecc = []

for i in range(50):
    prograde_ecc.append([])

for entry in dirs:
    if not entry.startswith("galaxy"):
        continue

    print(entry)

    for i in range(50):
        complete_path = os.path.join(file_path, entry, "galaxy_state_1_to_2")

        population_cabinet = snapshot_handler.load_cabinet(complete_path, f"galaxy_state_1_to_2_T{i}_blackholes")

        prograde_ecc[i].append(population_cabinet["prograde"]["orb_ecc"])
```

```python
ecc_v_time = [None] * 50

for i, entry in enumerate(prograde_ecc):
    combined = []

    for item in entry:
        combined.extend(item)

    ecc_v_time[i] = combined
```

```python
a = np.array([])
b = np.array([])
c = np.array([])
d = np.array([])

for time_step in ecc_v_time:
    a = np.append(a, sum([1 for x in time_step if x <= 0.01]))
    b = np.append(b, sum([1 for x in time_step if (0.01 < x <= 0.03)]))
    c = np.append(c, sum([1 for x in time_step if (0.03 < x <= 0.07)]))
    d = np.append(d, sum([1 for x in time_step if x > 0.07]))
```

```python
cats = range(50)

plt.bar(cats, a, label="ecc <= 0.01")
plt.bar(cats, b, bottom=a, label="0.01 < ecc <= 0.03")
plt.bar(cats, c, bottom=a+b, label="0.03 < ecc <= 0.07")
plt.bar(cats, d, bottom=a+b+c, label="ecc > 0.07")

plt.legend()
```

```python
from pathlib import Path

settings = SettingsManager({
    "verbose": False,
    "override_files": True,
    "save_state": True,
    "save_each_timestep": True
})

file_path = "../runs"
plots_dir = "../runs/plots"

snapshot_handler = TxtSnapshotHandler(settings)

dirs = os.listdir(file_path)

prograde_ecc = []

for i in range(50):
    prograde_ecc.append(np.array([]))

for entry in dirs:
    if not entry.startswith("gal"):
        continue

    print(entry)

    for i in range(50):
        complete_path = os.path.join(file_path, entry)

        data = np.loadtxt(Path(complete_path, f"output_bh_single_pro_{i}.dat"))

        prograde_ecc[i] = np.append(prograde_ecc[i], data[:, 6])

        print(len(prograde_ecc[i]))
```

```python
a = np.array([])
b = np.array([])
c = np.array([])
d = np.array([])

for time_step in prograde_ecc:
    a = np.append(a, sum([1 for x in time_step if x <= 0.01]))
    b = np.append(b, sum([1 for x in time_step if (0.01 < x <= 0.03)]))
    c = np.append(c, sum([1 for x in time_step if (0.03 < x <= 0.07)]))
    d = np.append(d, sum([1 for x in time_step if x > 0.07]))

cats = range(50)

plt.bar(cats, a, label="ecc <= 0.01")
plt.bar(cats, b, bottom=a, label="0.01 < ecc <= 0.03")
plt.bar(cats, c, bottom=a+b, label="0.03 < ecc <= 0.07")
plt.bar(cats, d, bottom=a+b+c, label="ecc > 0.07")

plt.legend()
```
